```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

Load the Dataset

```python
# Load the dataset
df = pd.read_csv('insurance.csv')

# Display the first few rows
df.head()
```

Scatter Plot: Age vs Charges

```python
# Scatter Plot of Age vs Chargesd
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['charges'], color='b')
plt.title('Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

Fit a Linear Regression Model

```python
# Define X and Y
x = df[['age']]
y = df[['charges']]

# Create and fit the model
model = LinearRegression()
model.fit(x,y)

# Make predictions
predictions = model.predict(x)
```

Plot with Best-Fit line

```python
# Scatter Plot and Regression line
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['charges'], color='b', label='Actual Data')
plt.plot(df['age'], predictions, color='red', label='Best-Fit Line')
plt.title('Linear Regression: Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend()
plt.grid(True)
plt.show()
```
